```python
def train_test_split(X, y, seed):
        np.random.seed(seed)
        p = np.random.permutation(X.shape[0])
        X = X[p,:]
        y = y[p]

        m = X.shape[0]
        print("total rows: ", m)
        training_size = int(m *0.67)

        X_train = X[:training_size, :]
        X_test = X[ training_size:, :]

        y_train = y[:training_size]
        y_test = y[training_size:]

        return X_train, X_test, y_train, y_test
```

```python
class NaiveBayesClassifier:
    def fit(self, X_train, y_train):
        # Calculate the prior probability of each class
        self.classes, counts = np.unique(y_train, return_counts=True)
        self.prior_probs = counts / len(y_train)
        
        # Calculate the conditional probability of each feature given each class
        self.cond_probs = {}
        for c in self.classes:
            X_c = X_train[y_train == c]
            self.cond_probs[c] = {
                "mean": np.mean(X_c, axis=0),
                "std": np.std(X_c, axis=0) + 1e-8  # Added  small epsilon to avoid division by zero
            }
    def predict(self, X_test):
        # Step 3: Predict the class
        y_pred = []
        for x in X_test:
            probs = []
            for c in self.classes:
                prior_prob = np.log(self.prior_probs[c])
                cond_prob = np.sum(np.log(self._gaussian_pdf(x, self.cond_probs[c]["mean"], self.cond_probs[c]["std"])))
                probs.append(prior_prob + cond_prob)
            y_pred.append(self.classes[np.argmax(probs)])
        return np.array(y_pred)
    
    def _gaussian_pdf(self, x, mean, std):
        # Gaussian Probability Density Function (PDF)
        exponent = -0.5 * ((x - mean) / std) ** 2
        return (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(exponent)

    def accuracy(self, X_test, y_test):
        y_pred = self.predict(X_test)
        accuracy = np.mean(y_pred == y_test)
        return accuracy
    def evaluate(self, X_test, y_test):
        y_pred = self.predict(X_test)
        
        # Calculate accuracy
        accuracy = np.mean(y_pred == y_test)
        
        # Calculate precision, recall, and F1-score
        tp = np.sum((y_pred == 1) & (y_test == 1))
        fp = np.sum((y_pred == 1) & (y_test == 0))
        fn = np.sum((y_pred == 0) & (y_test == 1))
        
        precision = tp / (tp + fp + 1e-8)  # Add small epsilon to avoid division by zero
        recall = tp / (tp + fn + 1e-8)
        f1_score = 2 * (precision * recall) / (precision + recall + 1e-8)
        
        print("accuracy: ", accuracy)
        print("precision: ", precision)
        print("recall: ", recall)
        print("F1 score: ", f1_score)
        
        return accuracy, precision, recall, f1_score
```
